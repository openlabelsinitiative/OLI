### Get trusted labels

````python
# pick a trust list, then set the source to the trust node (or create your own trust list: https://github.com/openlabelsinitiative/OLI/tree/main/3_label_trust)
# returns all trust lists: oli.trust.raw_trust_lists
oli.set_trust_node(address="0xa725646c05e6bb813d98c5abb4e72df4bcf00b56")
````

````python
# now you get a confidence score for each label e.g. 'confidence': 0.8
oli.get_trusted_labels(address="0x9438b8B447179740cD97869997a2FCc9b4AA63a2")
````

### Download full parquet exports

Raw: (updated daily)
```
https://api.growthepie.xyz/v1/oli/labels_raw.parquet
```
Decoded: (updated daily)
```
https://api.growthepie.xyz/v1/oli/labels_decoded.parquet
```

````python
oli.get_full_raw_export_parquet()
````

````python
oli.get_full_decoded_export_parquet()
````

````python
import pandas as pd

# download the decoded parquet file into a df
df = pd.read_parquet(oli.get_full_decoded_export_parquet())
df
````

````python
# pivot decoded_df by tag_id, tag_value and save into labels_df
df_pivot = df.pivot(index=['id', 'chain_id', 'address', 'attester', 'time_created', 'revoked', 'is_offchain'], columns='tag_id', values='tag_value')
df_pivot = df_pivot.reset_index()
df_pivot
````
